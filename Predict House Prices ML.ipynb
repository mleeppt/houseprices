```python
import pandas as pd

zillow_files = ["Metro_median_sale_price_uc_sfrcondo_week.csv", "Metro_median_sale_price_uc_sfrcondo_week.csv"]

dfs = [pd.read_csv(f) for f in zillow_files] 

dfs[0]
```

```python
dfs = [pd.DataFrame(df.iloc[0,5:]) for df in dfs]
```

```python
dfs[0]
```

```python
dfs[1]
```

```python
for df in dfs:
    df.index = pd.to_datetime(df.index)
    df["month"] = df.index.to_period("M")
dfs[1]
```

```python
price_data = dfs[0].merge(dfs[1], on="month")
price_data.index = dfs[0].index
price_data
```

#deleting month column

```python
del price_data["month"]
price_data.columns = ["price", "value"]
price_data
```

```python
fed_data = fed_data.dropna()
fed_data.tail(20)
```

```python
from datetime import timedelta
fed_data.index = fed_data.index + timedelta(days=1)
fed_data
```

```python
price_data = fed_data.merge(price_data, left_index=True, right_index=True)
price_data
```
